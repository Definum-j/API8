```python
def get_sum_odd(a, b):
    sum = 0 
    for i in range(a,b+1):
      if i % 2 == 1:
        sum = sum + i
    return sum
get_sum_odd(3,8)
    
```

```python
def grade(score):
  if score >= 0 and score < 60:
    return 'неудовлетворительно'
  if score >= 60 and score < 75:
    return 'удовлетворительно'
  if score >= 75 and score < 91:
    return 'хорошо'
  if score >= 91 and score < 101:
    return 'отлично'

a = grade(75)
print(a)
```

```python
def gen_list (a, b, c):
  numbers = []
  for i in range(a, b,c):
    numbers.append(i)
  return numbers

print(gen_list(30, 100, 33))
print(gen_list(1, 10, 2))
```

```python
def gen_dict(Name, Age):
  sample_dict = {}

  for i in range(0, len(Name)):
    sample_dict[Name[i]] = Age[i]
  
  return sample_dict
gen_dict(['Kate', 'Lena', 'Dima'], [23,56,33])
```
